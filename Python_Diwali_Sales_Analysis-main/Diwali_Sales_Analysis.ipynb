```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_df = pd.read_csv("Diwali Sales Data.csv", encoding= 'unicode_escape')
```

```python
data_df
```

```python
data_df.isna().sum()
```

```python
data_df.dropna(inplace = True)
```

```python
data_df.isna().sum()
```

```python
data_df.dtypes
```

```python
data_df.drop(['Status','unnamed1'], axis = 1, inplace= True)
```

```python
data_df
```

```python
data_df["Amount"].sum()
```

```python
data_df["Amount"] = data_df["Amount"].astype('int32')
```

```python
data_df["Amount"].sum()
```

```python
data_df.dtypes
```

```python
data_df.rename(columns = {"Marital_Status":"Shaadi"})
```

```python
data_df["Sales"] = data_df["Orders"] * data_df["Amount"]
```

```python
data_df
```

```python
ax = sns.countplot(x = 'Zone',data = data_df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = data_df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data_df
```

```python
data_df['Marital_Status'] = data_df['Marital_Status'].replace({0: 'no', 1: 'yes'})
```

```python
data_df
```

```python
product_sales = data_df.groupby("Product_Category")[["Sales"]].sum()
```

```python
product_sales
```

```python
products = data_df["Product_Category"].unique()
```

```python
plt.figure(figsize=(20,5))
plt.xticks(rotation=45)
plt.grid()
plt.ticklabel_format(style='plain', useOffset=False)
plt.bar(products,product_sales["Sales"])
```
